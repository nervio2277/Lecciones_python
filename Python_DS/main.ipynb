```python
dict_country_leagues     = {
'Spanish La Liga': 'SP1', 'Spanish Segunda Division': 'SP2',
'German Bundesliga': 'D1',
'English Premier League': 'E0', 'English League 1': 'E2', 'English League 3': 'E3',
'English Lueague Conference':'EC'}
```

```python
dict_country_leagues['Spanish La Liga']
    
```

```python
for league in dict_country_leagues:
    print(dict_country_leagues[league])
```

```python
dict_hist_data = {}

#Hacer lo mismo pero con multiples temporadas
#leagues = ['E0', 'E1', 'E2', 'E3', 'EC']

for league in dict_country_leagues:
    frames=[]
    for temporada in range(15,21):
        print("descargando ..."+root_url+ str(temporada)+str(temporada+1)+"/" + dict_country_leagues[league] + ".csv")
        df = pd.read_csv(root_url+ str(temporada)+str(temporada+1)+"/" + dict_country_leagues[league] + ".csv",encoding='unicode-escape')
        df.insert(1, "Temporada", temporada)
        frames.append(df)
    df_concat =pd.concat(frames)
    dict_hist_data[league] = df_concat
```
